# NFT
NFT는 게임의 코인을 비트코인과 같이 가상화폐로 사용할 수 있는데 엔씨소프트에서도 이 NFT를 적용한다고 하였기때문에 NFT의 검색량이 많을수록 엔씨로프트 주가는 상승할 것이라고 생각했습니다.
